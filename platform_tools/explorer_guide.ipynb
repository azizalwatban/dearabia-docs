# Using the Metadata Explorer

How do we know specifically **where** and **when** data is available? Before we start any analysis, we can answer this question by verifying existing data on the [DE Arabia Metadata Explorer](https://explorer.digitaleartharabia.ssa.gov.sa/products).

## Map or Explorer?

The [DE Arabia Metadata Explorer](https://explorer.digitaleartharabia.ssa.gov.sa/products) and [DE Arabia Map](https://digitaleartharabia.ssa.gov.sa/index.html) look similar, but they are designed for different purposes. 

Use the DE Arabia Map if you:

* Want to see what the product or dataset looks like and interact with the data in a map viewer display

Use the DE Arabia Metadata Explorer if you:

* Want to know exactly where and when you can find data

## Open the DE Arabia Metadata Explorer

1. Open [https://explorer.digitaleartharabia.ssa.gov.sa/products](https://explorer.digitaleartharabia.ssa.gov.sa/products). This will display the Metadata Explorer user interface.

<img align="middle" src="../_static/platform_tools/explorer/NumberingGuide.png" alt="The DE Africa Metadata Explorer." width=600>

The Digital Earth Arabia Metadata Explorer has four main sections.

1. **Product selection:** This shows the currently-selected product. Click the selected product name to open the products dropdown menu.
2. **Time period:** This shows the time period for which the selected product is being displayed. Click the selected time period to open the time selection dropdown menu. The blue bar chart shows the data availability over time for the product you have selected.
3. **Map display:** This shows where data is available, for the selected time and product. Blue shaded tiles indicate the presence of data.
4. **Product information:** The sidebar shows more information about the data for the selected time and product. For example, this includes how many datasets are selected, the name of the datasets in the product, and its coordinate reference system.

### Select the Landsat 9 product

1. Click the **product selection** bar to open the dropdown menu. Select **landsat_9_sr**. This selects the Landsat 9 product.

<img align="middle" src="../_static/platform_tools/explorer/SelectLandsat.png" alt="Metadata Explorer Landsat 8." width=600>

2. Click the **time** bar to open the dropdown menu. Select **2025**. This will show all Landsat 9 datasets for 2025.

<img align="middle" src="../_static/platform_tools/explorer/Select2025" alt="Metadata Explorer Landsat 8 2018." width=600>

3. Click the **all months** bar to open the dropdown menu. Select **August**. This will show all the Landsat 9 datasets for August 2025.

<img align="middle" src="../_static/platform_tools/explorer/SelectAugust.png" alt="Metadata Explorer Landsat 8 2018." width=600>

4. Click the **all days** bar to open the dropdown menu. Select **6th**. This will show all the Landsat 9 datasets for 06 August 2025.

<img align="middle" src="../_static/platform_tools/explorer/Select6th.png" alt="Metadata Explorer Landsat 8 2018." width=600>

5. The **map display** will now show all the Landsat 9 datasets for 06 August 2025 as blue shaded boxes. Use the **+** button on the map to zoom in, and click and drag to pan the map. 

<img align="middle" src="../_static/platform_tools/explorer/ViewStrip.png" alt="Metadata Explorer Landsat 8 2018." width=600>

We can see that there is only data available for some parts of Saudi Arabia. Let's take a closer look at Riyadh.

### Zoom in on Riyadh

Use the map's **+** button to zoom in on Riyadh.

<img align="middle" src="../_static/platform_tools/explorer/Riyadh.png" alt="Metadata Explorer Landsat 8 2018." width=600>

When we hover over Riyadh, we can see there is one dataset for Riyadh for the 06 August 2025. Click on the dataset on the map. We can now see a preview of the data for that tile, as well as all the metadata associated with that dataset below the map viewer.
